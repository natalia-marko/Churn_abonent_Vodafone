## defining the functions

### def function adding_to_model_results

```python
model_results_df = pd.DataFrame(columns=['train_auc', 'test_auc', 'train_F1', 'test_F1', 'elapsed_time', '_auc', '_F1'])

def adding_to_model_results(existing_df, model_results, model_name):

    model_name = f"{model_name}_{pd.Timestamp.now().strftime('%Y-%m-%d %H:%M')}"

    # Create DataFrame from results
    new_df = pd.DataFrame(model_results, index=[model_name])

    # Convert string values to float
    new_df['train_auc'] = new_df['train_auc'].astype(float)
    new_df['test_auc'] = new_df['test_auc'].astype(float)
    new_df['train_F1'] = new_df['train_F1'].astype(float)
    new_df['test_F1'] = new_df['test_F1'].astype(float)

    # Calculate delta AUC and F1
    new_df['_auc'] = new_df['train_auc'] - new_df['test_auc']
    new_df['_F1'] = new_df['train_F1'] - new_df['test_F1']

    # Append new_df to the existing DataFrame
    updated_df = pd.concat([existing_df, new_df])

    return updated_df

```

### def function TRAIN

```python
def train_catboost_model(cb_model, X_train, y_train, X_test, y_test, iterations=None, params=None, optimal_threshold=None):
    start = time.time()

    evals_result = {}

    cb_model_results = {
        'train_auc': [],
        'test_auc': [],
        'train_F1': [],
        'test_F1': [],
        'test_TP': [],
        'test_FP': [],
        'test_FN': [],
        'elapsed_time': []
    }
    default_params = {'iterations': iterations,
                  'loss_function': 'Logloss',
                  'eval_metric': 'AUC',
                  'auto_class_weights': 'Balanced',
                  'random_seed': 42,
                  'verbose': 50}
    if params:
        default_params.update(params)

    cb_model.set_params(**default_params)

    cb_model.fit(X_train, y_train, eval_set=(X_test, y_test), use_best_model=True, plot=True)

    cb_train_pred_proba = cb_model.predict_proba(X_train)[:, 1]
    cb_pred_proba = cb_model.predict_proba(X_test)[:, 1]

    if optimal_threshold:
        cb_train_pred = [1 if proba > optimal_threshold else 0 for proba in cb_train_pred_proba]
        cb_pred = [1 if proba > optimal_threshold else 0 for proba in cb_pred_proba]
    else:
        cb_train_pred = cb_model.predict(X_train)
        cb_pred = cb_model.predict(X_test)

    cb_train_auc = roc_auc_score(y_train, cb_train_pred_proba)
    cb_train_f1 = f1_score(y_train, cb_train_pred)
    cb_test_auc = roc_auc_score(y_test, cb_pred_proba)
    cb_test_f1 = f1_score(y_test, cb_pred)

    #      
    cb_model_results["train_auc"].append(f"{cb_train_auc:.4f}")
    cb_model_results["train_F1"].append(f"{cb_train_f1:.4f}")
    cb_model_results["test_auc"].append(f"{cb_test_auc:.4f}")
    cb_model_results["test_F1"].append(f"{cb_test_f1:.4f}")

    print(f"Train Auc: {cb_train_auc:.2f}, Test Auc: {cb_test_auc:.2f}, Train F1: {cb_train_f1:.2f}, Test F1: {cb_test_f1:.2f}")

    train_confusion = confusion_matrix(y_train, cb_train_pred)
    test_confusion = confusion_matrix(y_test, cb_pred)
    train_classification_rep = classification_report(y_train, cb_train_pred)
    test_classification_rep = classification_report(y_test, cb_pred)

    FP = test_confusion[0][1]
    FN = test_confusion[1][0]
    TP = test_confusion[1][1]
    cb_model_results["test_FP"].append(FP)
    cb_model_results["test_FN"].append(FN)
    cb_model_results["test_TP"].append(TP)

    evals_result = cb_model.get_evals_result()
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

    #      
    im1 = ax1.imshow(train_confusion, interpolation='nearest', cmap='Blues')
    ax1.set_xticks(np.arange(train_confusion.shape[1]))
    ax1.set_yticks(np.arange(train_confusion.shape[0]))
    ax1.set_xlabel('Predicted labels')
    ax1.set_ylabel('True labels')
    ax1.set_title('Train Confusion Matrix')
    ax1.set_xticklabels(['0', '1'])
    ax1.set_yticklabels(['0', '1'])
    #     
    for i in range(train_confusion.shape[0]):
        for j in range(train_confusion.shape[1]):
          text = ax1.text(j, i, train_confusion[i, j], ha='center', va='center', color='black')
    plt.colorbar(im1, ax=ax1)

    #      
    im2 = ax2.imshow(test_confusion, interpolation='nearest', cmap='Blues')
    ax2.set_xticks(np.arange(test_confusion.shape[1]))
    ax2.set_yticks(np.arange(test_confusion.shape[0]))
    ax2.set_xlabel('Predicted labels')
    ax2.set_ylabel('True labels')
    ax2.set_title('Test Confusion Matrix')
    ax2.set_xticklabels(['0', '1'])
    ax2.set_yticklabels(['0', '1'])
    for i in range(test_confusion.shape[0]):
        for j in range(test_confusion.shape[1]):
            text = ax2.text(j, i, test_confusion[i, j], ha='center', va='center', color='black')
    plt.colorbar(im2, ax=ax2)

    plt.tight_layout()
    plt.show()

    print('\nTrain AUC-ROC:', cb_train_auc)
    print('\nTest AUC-ROC:', cb_test_auc)
    # print('\nTrain Confusion Matrix:\n', train_confusion)
    # print('\nTest Confusion Matrix:\n', test_confusion)
    print('\nTrain Classification Report:\n', train_classification_rep)
    print('\nTest Classification Report:\n', test_classification_rep)

    params_state = cb_model.get_params()
    for param, value in params_state.items():
        print(f"{param}: {value}")

    end = time.time()
    cb_elapsed_time = end - start

    cb_model_results["elapsed_time"].append(f"{cb_elapsed_time:.2f}")

    print(f" : {cb_elapsed_time:.2f} ")


    return cb_model, cb_model_results, cb_train_pred_proba, cb_pred_proba, evals_result
```

### def PLOT function

```python
import matplotlib.pyplot as plt

def plot_metrics(evals_result, figsize=(10,5)):
    #  
    train_logloss = evals_result['learn']['Logloss']
    test_logloss = evals_result['validation']['Logloss']
    test_AUC = evals_result['validation']['AUC']

    #   Logloss
    plt.figure(figsize=figsize)
    plt.plot(train_logloss, label='Train Logloss')
    plt.plot(test_logloss, label='Test Logloss')
    plt.xlabel('Iteration')
    plt.ylabel('Logloss')
    plt.title('Train and Test Logloss over Iterations')
    plt.legend()
    plt.grid(True)
    plt.show()

    #   AUC
    plt.figure(figsize=figsize)
    plt.plot(test_AUC, label='Test AUC')
    plt.xlabel('Iteration')
    plt.ylabel('AUC')
    plt.title('AUC Metric Over Iterations')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import seaborn as sns

def plot_proba_distribution(train_proba, test_proba):
    plt.figure(figsize=(10, 6))

    # Visualize the probability distribution for the training data
    sns.kdeplot(train_proba, color='blue', label='Training data')

    # Visualize the probability distribution for the test data
    sns.kdeplot(test_proba, color='red', label='Test data')

    plt.title('Probability distribution for training and test data')
    plt.xlabel('Predicted probability')
    plt.ylabel('Density')
    plt.legend()

    plt.show()
```

```python
def train_catboost_model_f1(cb_model, X_train, y_train, X_test, y_test, iterations=None, params=None, optimal_threshold=None):
    start = time.time()

    cb_model_results = {
        'train_precision': [],
        'test_precision': [],
        'train_recall': [],
        'test_recall': [],
        'train_F1': [],
        'test_F1': [],
        'test_TP': [],
        'test_FP': [],
        'test_FN': [],
        'elapsed_time': []
    }
    default_params = {'iterations': iterations,
                  'loss_function': 'Logloss',
                  'auto_class_weights': 'Balanced',
                  'random_seed': 42,
                  'verbose': 50}
    if params:
        default_params.update(params)

    cb_model.set_params(**default_params)

    cb_model.fit(X_train, y_train, eval_set=(X_test, y_test), use_best_model=True, plot=True)

    cb_train_pred_proba = cb_model.predict_proba(X_train)[:, 1]
    cb_pred_proba = cb_model.predict_proba(X_test)[:, 1]

    if optimal_threshold:
        cb_train_pred = [1 if proba > optimal_threshold else 0 for proba in cb_train_pred_proba]
        cb_pred = [1 if proba > optimal_threshold else 0 for proba in cb_pred_proba]
    else:
        cb_train_pred = cb_model.predict(X_train)
        cb_pred = cb_model.predict(X_test)

    cb_train_precision = precision_score(y_train, cb_train_pred)
    cb_train_recall = recall_score(y_train, cb_train_pred)
    cb_train_f1 = f1_score(y_train, cb_train_pred)
    cb_test_precision = precision_score(y_test, cb_pred)
    cb_test_recall = recall_score(y_test, cb_pred)
    cb_test_f1 = f1_score(y_test, cb_pred)

    #      
    cb_model_results["train_precision"].append(f"{cb_train_precision:.4f}")
    cb_model_results["train_recall"].append(f"{cb_train_recall:.4f}")
    cb_model_results["train_F1"].append(f"{cb_train_f1:.4f}")
    cb_model_results["test_precision"].append(f"{cb_test_precision:.4f}")
    cb_model_results["test_recall"].append(f"{cb_test_recall:.4f}")
    cb_model_results["test_F1"].append(f"{cb_test_f1:.4f}")

    print(f"Train Precision: {cb_train_precision:.2f}, Test Precision: {cb_test_precision:.2f}, Train Recall: {cb_train_recall:.2f}, Test Recall: {cb_test_recall:.2f}, Train F1: {cb_train_f1:.2f}, Test F1: {cb_test_f1:.2f}")

    train_confusion = confusion_matrix(y_train, cb_train_pred)
    test_confusion = confusion_matrix(y_test, cb_pred)
    train_classification_rep = classification_report(y_train, cb_train_pred)
    test_classification_rep = classification_report(y_test, cb_pred)

    FP = test_confusion[0][1]
    FN = test_confusion[1][0]
    TP = test_confusion[1][1]
    cb_model_results["test_FP"].append(FP)
    cb_model_results["test_FN"].append(FN)
    cb_model_results["test_TP"].append(TP)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

    #    ...

    params_state = cb_model.get_params()
    for param, value in params_state.items():
        print(f"{param}: {value}")

    end = time.time()
    cb_elapsed_time = end - start

    cb_model_results["elapsed_time"].append(f"{cb_elapsed_time:.2f}")

    print(f" : {cb_elapsed_time:.2f} ")

    return cb_model, cb_model_results, cb_train_pred_proba, cb_pred_proba
```

```python
test_data.drop('abon_id',axis=1, inplace=True)

#       
X_test = test_data.drop('target', axis=1)
y_test = test_data['target']
```

### testing and geting evaluation metrics

```python
models = [cb_model_0, cb_model_1, cb_model_fea]
model_names = ['cb_model_0', 'cb_model_1', 'cb_model_fea']
```

```python
#  DataFrame   
results = pd.DataFrame(columns=['Model', 'AUC', 'F1 Score', 'Training time', 'TP', 'FP', 'FN'])

for i, model in enumerate(models):
    start_time = time.time()
    predictions_proba = model.predict_proba(X_test)[:, 1]
    predictions = model.predict(X_test)
    end_time = time.time()
    training_time = end_time - start_time

    auc = roc_auc_score(y_test, predictions_proba)
    f1 = f1_score(y_test, predictions)

    cm = confusion_matrix(y_test, predictions)
    fp = cm[0][1]
    fn = cm[1][0]
    tp = cm[1][1]

    #    DataFrame
    results = results.append({'Model': model_names[i],
                              'AUC': auc,
                              'F1 Score': f1,
                              'TP': tp,
                              'FP': fp,
                              'FN': fn,
                              'Training time': training_time},
                              ignore_index=True)

    #       
    print(f'\nResults for model: {model_names[i]}')
    print('----------------------------')
    print('Classification Report:')
    print(classification_report(y_test, predictions))
    print('----------------------------\n')
    plt.figure(figsize=(5,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'\nConfusion matrix for model: {model_names[i]}')
    plt.show()
    optimal_threshold, max_f1 = my_functions.calculate_optimal_threshold(y_test, predictions_proba)
    print('----------------------------')
    print(f'\nOptimal threshold for model: {model_names[i]} is:{optimal_threshold}')

#  
print('Summary of model performance:')
print(results)
```
