```python
model_results_df = pd.DataFrame(columns=['train_auc', 'test_auc', 'train_F1', 'test_F1', 'elapsed_time', 'Δ_auc', 'Δ_F1'])
```

```python
def adding_to_model_results(existing_df, model_results, model_name):

    model_name = f"{model_name}_{pd.Timestamp.now()}"

    # Create DataFrame from results
    new_df = pd.DataFrame(model_results, index=[model_name])

    # Convert string values to float
    new_df['train_auc'] = new_df['train_auc'].astype(float)
    new_df['test_auc'] = new_df['test_auc'].astype(float)
    new_df['train_F1'] = new_df['train_F1'].astype(float)
    new_df['test_F1'] = new_df['test_F1'].astype(float)

    # Calculate delta AUC and F1
    new_df['Δ_auc'] = new_df['train_auc'] - new_df['test_auc']
    new_df['Δ_F1'] = new_df['train_F1'] - new_df['test_F1']

    # Append new_df to the existing DataFrame
    updated_df = pd.concat([existing_df, new_df])

    return updated_df
```

#### def function TRAIN

```python
def train_catboost_model(cb_model, X_train, y_train, X_test, y_test, iterations=None, params=None, optimal_threshold=None):
    start = time.time()

    evals_result = {}

    cb_model_results = {
        'train_auc': [],
        'test_auc': [],
        'train_F1': [],
        'test_F1': [],
        'elapsed_time': []
    }
    default_params = {'iterations': iterations,
                  # 'loss_function': 'Logloss',
                  'eval_metric': 'AUC',
                  'auto_class_weights': 'Balanced',
                  'random_seed': 42,
                  'verbose': 50}
    if params:
        default_params.update(params)

    cb_model.set_params(**default_params)  # apply parameters to the model

    cb_model.fit(X_train, y_train, eval_set=(X_test, y_test), use_best_model=True, plot=True)

    cb_train_pred_proba = cb_model.predict_proba(X_train)[:, 1]
    cb_pred_proba = cb_model.predict_proba(X_test)[:, 1]

    if optimal_threshold:
        cb_train_pred = [1 if proba > optimal_threshold else 0 for proba in cb_train_pred_proba]
        cb_pred = [1 if proba > optimal_threshold else 0 for proba in cb_pred_proba]
    else:
        cb_train_pred = cb_model.predict(X_train)
        cb_pred = cb_model.predict(X_test)

    cb_train_auc = roc_auc_score(y_train, cb_train_pred_proba)
    cb_train_f1 = f1_score(y_train, cb_train_pred)
    cb_test_auc = roc_auc_score(y_test, cb_pred_proba)
    cb_test_f1 = f1_score(y_test, cb_pred)

    # Додайте результати до тренувального набору даних
    cb_model_results["train_auc"].append(f"{cb_train_auc:.4f}")
    cb_model_results["train_F1"].append(f"{cb_train_f1:.4f}")
    cb_model_results["test_auc"].append(f"{cb_test_auc:.4f}")
    cb_model_results["test_F1"].append(f"{cb_test_f1:.4f}")

    print(f"Train Auc: {cb_train_auc:.2f}, Test Auc: {cb_test_auc:.2f}, Train F1: {cb_train_f1:.2f}, Test F1: {cb_test_f1:.2f}")

    train_confusion = confusion_matrix(y_train, cb_train_pred)
    test_confusion = confusion_matrix(y_test, cb_pred)
    train_classification_rep = classification_report(y_train, cb_train_pred)
    test_classification_rep = classification_report(y_test, cb_pred)

    evals_result = cb_model.get_evals_result()

    print('\nTrain AUC-ROC:', cb_train_auc)
    print('\nTest AUC-ROC:', cb_test_auc)
    print('\nTrain Confusion Matrix:\n', train_confusion)
    print('\nTest Confusion Matrix:\n', test_confusion)
    print('\nTrain Classification Report:\n', train_classification_rep)
    print('\nTest Classification Report:\n', test_classification_rep)

    params_state = cb_model.get_params()
    for param, value in params_state.items():
        print(f"{param}: {value}")

    end = time.time()
    cb_elapsed_time = end - start

    cb_model_results["elapsed_time"].append(f"{cb_elapsed_time:.2f}")

    print(f"Час виконання: {cb_elapsed_time:.2f} сек")


    return cb_model, cb_model_results, cb_train_pred_proba, cb_pred_proba, evals_result
```

#### def PLOT function

```python
import matplotlib.pyplot as plt

def plot_metrics(evals_result, figsize=(10,5)):
    # Розпаковуємо результати
    train_logloss = evals_result['learn']['Logloss']
    test_logloss = evals_result['validation']['Logloss']
    test_AUC = evals_result['validation']['AUC']

    # Графік для Logloss
    plt.figure(figsize=figsize)
    plt.plot(train_logloss, label='Train Logloss')
    plt.plot(test_logloss, label='Test Logloss')
    plt.xlabel('Iteration')
    plt.ylabel('Logloss')
    plt.title('Train and Test Logloss over Iterations')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Графік для AUC
    plt.figure(figsize=figsize)
    plt.plot(test_AUC, label='Test AUC')
    plt.xlabel('Iteration')
    plt.ylabel('AUC')
    plt.title('AUC Metric Over Iterations')
    plt.legend()
    plt.grid(True)
    plt.show()
```

#### def PLOT PROBA function

```python
def plot_proba_distribution(train_proba, test_proba):
    plt.figure(figsize=(10, 6))

    # Visualize the probability distribution for the training data
    sns.kdeplot(train_proba, color='blue', label='Training data')

    # Visualize the probability distribution for the test data
    sns.kdeplot(test_proba, color='red', label='Test data')

    plt.title('Probability distribution for training and test data')
    plt.xlabel('Predicted probability')
    plt.ylabel('Density')
    plt.legend()

    plt.show()
```

```python
test_data = my_functions.load('/content/drive/My Drive/Colab Notebooks/churn_test_model_fe.pickle')
```

```python
test_data.drop('abon_id',axis=1, inplace=True)

# Розділяємо дані на ознаки та цільову змінну
X_test = test_data.drop('target', axis=1)
y_test = test_data['target']

models = [cb_model_0, cb_model_1, cb_model_2, cb_model_3]
model_names = ['cb_model_0', 'cb_model_1', 'cb_model_2', 'cb_model_3']
# Створюємо DataFrame для зберігання результатів
results = pd.DataFrame(columns=['Model', 'AUC', 'F1 Score', 'Training time', 'False Positives', 'False Negatives'])

for i, model in enumerate(models):
    start_time = time.time()
    predictions_proba = model.predict_proba(X_test)[:, 1]
    predictions = model.predict(X_test)
    end_time = time.time()
    training_time = end_time - start_time

    auc = roc_auc_score(y_test, predictions_proba)
    f1 = f1_score(y_test, predictions)

    cm = confusion_matrix(y_test, predictions)
    fp = cm[0][1]
    fn = cm[1][0]

    # Додаємо результати до DataFrame
    results = results.append({'Model': model_names[i],
                              'AUC': auc,
                              'F1 Score': f1,
                              'False Positives': fp,
                              'False Negatives': fn,
                              'Training time': training_time},
                              ignore_index=True)

    # Друкуємо повідомлення з результатами для поточної моделі
    print(f'\nResults for model: {model_names[i]}')
    print('----------------------------')
    print('Classification Report:')
    print(classification_report(y_test, predictions))
    print('----------------------------\n')

# Показуємо результати
print('Summary of model performance:')
print(results)
```

В цьому контексті:

False Positives: Це ситуації, коли модель помилково визначає, що клієнт залишиться, хоча насправді він має намір піти. Це може призвести до втрати клієнта, оскільки ви не прийняли вчасних заходів, щоб утримати його.

False Negatives: Це ситуації, коли модель помилково визначає, що клієнт піде, хоча насправді він залишиться. Це може призвести до непотрібних витрат на утримання клієнта, який і так не мав наміру піти.

В залежності від вашої бізнес-стратегії, може бути важливо зосередитись на мінімізації одного з цих видів помилок. Наприклад, якщо ви вважаєте втрату клієнта більш критичною, ви можете прагнути знизити кількість помилок False Positives.


Так, згідно з вашим контекстом, False Negatives важливі, оскільки вони означають, що ви можливо пропустите клієнтів, які мають намір покинути ваш сервіс. Це означає, що ви можливо не приймете вчасних заходів, щоб утримати цих клієнтів. Тому, вибираючи модель, вам слід звернути особливу увагу на те, яка модель має найменшу кількість помилок False Negatives.

Однак важливо зазначити, що оптимальний вибір моделі залежить від більшої кількості факторів, ніж просто мінімізація False Negatives. Ви також повинні врахувати інші метрики якості моделі, такі як точність, повноту, F1-оцінку та AUC-ROC. Крім того, важливо враховувати бізнес-вартість False Positives і False Negatives, а також потенційні витрати на утримання або втрату клієнтів.

Згідно з даними, які ви надали, моделі `cb_model_2` та `cb_model_3` мають найнижчу кількість помилок False Negatives, що означає, що вони краще за інші моделі виявляють клієнтів, які планують покинути ваш сервіс.

Однак, перед тим як вирішити, яку модель вибрати, ви також повинні врахувати інші метрики моделей. Наприклад, `cb_model_3` має найнижчий F1 Score, що означає, що вона може не бути найкращою з точки зору точності та повноти. Також `cb_model_3` має вищу кількість помилок False Positives, що може призвести до непотрібних витрат на утримання клієнтів, які не планують покинути сервіс.

Враховуючи ці фактори, вам слід вивчити бізнес-вартість кожного виду помилок і вирішити, яка модель найкраще відповідає вашим бізнес-цілям і вимогам. Це може означати вибір моделі, яка має більшу кількість помилок False Negatives, але краще впорається з іншими метриками.

```python
cb_2_pred_proba = cb_model_2.predict_proba(X_test)[:, 1]
optimal_threshold, max_f1 = my_functions.calculate_optimal_threshold(y_test, cb_2_pred_proba)
```

Якщо ви знайшли нове значення порогу, яке, може зменшити кількість помилок false negatives і підвищити загальну ефективність моделі, можемо застосувати цей новий поріг для прогнозів.


```python
threshold = 0.8

models = [cb_model_2]
model_names = ['cb_model_2']
# Створюємо DataFrame для зберігання результатів
results = pd.DataFrame(columns=['Model', 'AUC', 'F1 Score', 'Training time', 'False Positives', 'False Negatives'])

for i, model in enumerate(models):
    start_time = time.time()
    predictions_proba = model.predict_proba(X_test)[:, 1]
    predictions = (predictions_proba >= threshold).astype('int')
    end_time = time.time()
    training_time = end_time - start_time

    auc = roc_auc_score(y_test, predictions_proba)
    f1 = f1_score(y_test, predictions)

    cm = confusion_matrix(y_test, predictions)
    fp = cm[0][1]
    fn = cm[1][0]

    # Додаємо результати до DataFrame
    results = results.append({'Model': model_names[i],
                              'AUC': auc,
                              'F1 Score': f1,
                              'False Positives': fp,
                              'False Negatives': fn,
                              'Training time': training_time},
                              ignore_index=True)

    # Друкуємо повідомлення з результатами для поточної моделі
    print(f'\nResults for model: {model_names[i]}')
    print('----------------------------')
    print('Classification Report:')
    print(classification_report(y_test, predictions))
    print('----------------------------\n')

# Показуємо результати
print('Summary of model performance:')
print(results)
```
